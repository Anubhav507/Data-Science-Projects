## Predicting value of house

### Model 1


```python
plt.figure(figsize = (16, 12))
plt.subplot(2,3,1)
sns.scatterplot(x=df_sales['sqft_living'],y=df_sales['price'])

plt.subplot(2,3,2)
sns.scatterplot(x=df_sales['bathrooms'],y=df_sales['price'])

plt.subplot(2,3,3)
sns.scatterplot(x=df_sales['bedrooms'],y=df_sales['price'])

plt.subplot(2,3,4)
sns.scatterplot(x=df_sales['sqft_lot'],y=df_sales['price'])

plt.subplot(2,3,5)
sns.scatterplot(x=df_sales['floors'],y=df_sales['price'])

plt.subplot(2,3,6)
sns.scatterplot(x=df_sales['zipcode'],y=df_sales['price'])
plt.show()
```

```python
x=df_sales[['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'zipcode']]
y=df_sales[['price']]
```

```python
print(x)
```

we are not performing standardization or normalization as this is linear regression and we do that normally in clustering or ridge and lasso regression

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(xtrain,ytrain)
```

```python
y_hat=model.predict(xtest)
y_hat
```

```python
import sklearn
import math
mse = sklearn.metrics.mean_squared_error(ytest,y_hat)
rmse = math.sqrt(mse)
print(rmse)
```
